```python
avg_r=rain['avg_rain'].mean()
avg=temp['avg_temp'].mean()
diff=temp['avg_temp'].max()-temp['avg_temp'].min()
diff_r=rain['avg_rain'].max()-rain['avg_rain'].min()
```

```python
dev=[]
def deviation(data):
    for _,row in temp.iterrows():
        dev.append((avg-row['avg_temp'])/diff)
    return dev
temp['dev']=deviation(temp)
dev=[]
def deviation_r(data):
    for _,row in data.iterrows():
        dev.append((avg_r-row['avg_rain'])/diff_r)
    return dev
rain['dev']=deviation_r(rain)
```
